```python
url = "https://github.com/Azure/azure-cosmos-dotnet-v3/files/6006774/diagnostics.json.txt";

import requests
import json
response = json.loads(requests.get(url).text)


from anytree import Node, RenderTree

def tree_builder_short(d, parent, l=0):
    if isinstance(d, list):
        for i, item in enumerate(d):
            if "Value" in item:
                child = item['Value']
            else:
                child = item
            tree_builder_short(child, parent, l + 1)
    else:
        print("trace" + d['Name'] + "," + d['Duration'])
        node = Node(
            name   = d['Name'],
            key    = d['Duration'],
            parent = parent
        )
        if isinstance(d['Children'], list):        
            tree_builder_short(d['Children'], node, l + 1)
    

def tree_builder(d, parent, l=0):
    if isinstance(d, list):
        for i, item in enumerate(d):
            node_uid = 'l{}n{}'.format(l, i)
            print("arrayitem " + node_uid + ",")
            node = Node(
                name   = i,
                key    = node_uid,
                parent = parent,
                an_attr = "a"
            )
            tree_builder(item, node, l + 1)
    else:
        for i, (k, v) in enumerate(d.items()):
            node_uid = 'l{}n{}'.format(l, i)
            print("object prop " + node_uid + "," + k)
            node = Node(
                name   = k,
                key    = node_uid,
                parent = parent
            )
            if isinstance(v, dict):
                node.an_attr = ''
                tree_builder(v, node, l + 1)
            else:
                node.an_attr = v

root  = Node(name='root', key='root', an_attr='c')
tree_builder_short(response, root)
```

```python
for pre, fill, node in RenderTree(root):
    print("%s%s|%s" % (pre, node.name, node.key))
```
